```python
# gamma Li+, square root of the ionic strength (= concentration for a 1:1 electrolyte, m^(1/2)) , m= molaLity, mol/Kg
li_data=np.transpose(np.array([[0.0972,0.138,0.171,0.221,0.311,0.445,0.701,0.889,0.999,1.09],
                   [0.909,0.885,0.874,0.846,0.827,0.812,0.857,0.938,0.994,1.06]]))
li=np.array(li_data)

# gamma Cl-, square root of the ionic strength (= concentration for a 1:1 electrolyte, m^(1/2)) 
cl_data=np.transpose(np.array([[0.0972,0.138,0.171,0.221,0.311,0.445,0.701,0.889,0.999,1.09],
                   [0.901,0.866,0.835,0.810,0.755,0.706,0.639,0.613,0.608,0.603]]))
cl=np.array(cl_data)
```

```python
# Step 1

plt.figure(figsize=(6, 6), dpi= 80, facecolor='w', edgecolor='k')
plt.tick_params(axis='both', which='both', length=10.0, labelleft=True, labelbottom=True, labelsize=20.0)

plt.plot(li[:,0], li[:,1], 'o', markerfacecolor='green', markersize=12, markeredgecolor='blue', linewidth=2.5, label="Li+, Exp.") 
plt.plot(li[:,0], cl[:,1], 'o', markerfacecolor='blue', markersize=12, markeredgecolor='orange', linewidth=2.5, label="Cl-, Exp.")

plt.xlim(0,1.2)
# plt.ylim(-0.6,0.2)

plt.xlabel('$\sqrt{m}$', fontsize=26)
plt.ylabel('$\gamma_{Li^{+},Cl^{-}}$', fontsize=26)
plt.legend()
plt.savefig('foo.png')
plt.show()
```

```python
# Step 2

# Debye-Hückel - Electrochemistry
c=np.arange(0, 100, dtype=float) # concentration
nc=len(c) 

# Ionic strength, [mol/Kg]
ist=np.zeros(100) # Ionic strength
lg=np.zeros(100) # Ln (base e) of the mean activity
A=0.509 # Value of A tipically used for water at room temperature [Kg/mol]^{1/2} IF
        # we work with log_10 instead of log_e (in which case A=1.172 [Kg/mol]^{1/2})

for i in range(0,nc):
    ist[i]=c[i]*0.01 # Ionic strength = concentration (in this case!) [mol/Kg]^{1/2}
    lg[i]=-A*np.sqrt(ist[i]) # Log (base 10) of the mean activity coefficient
    
```

```python
plt.figure(figsize=(6, 6), dpi= 80, facecolor='w', edgecolor='k')
plt.tick_params(axis='both', which='both', length=10.0, labelleft=True, labelbottom=True, labelsize=20.0)

plt.plot(li[:,0], np.log10(li[:,1]), 'o', markerfacecolor='green', markersize=12, markeredgecolor='blue', linewidth=2.5, label="Li+, Exp.") 
plt.plot(li[:,0], np.log10(cl[:,1]), 'o', markerfacecolor='blue', markersize=12, markeredgecolor='orange', linewidth=2.5, label="Cl-, Exp.")
plt.plot((np.sqrt(ist[:])), (lg[:]), 'o', markerfacecolor='red', markersize=12, markeredgecolor='blue', 
         linewidth=2.5, label="DH")

plt.xlim(0,1.2)
# plt.ylim(-0.6,0.2)

plt.xlabel('$\sqrt{m}$', fontsize=26)
plt.ylabel('$log_{10} \gamma_{Li^{+},Cl^{-}}$', fontsize=26)
plt.legend()
plt.savefig('foo.png')
plt.show()
```

```python
# Step 4

A=0.51006 # (kg-solvent/mol-solute)1/2 -> From Table 2 of: J. Phys. Chem. B 2012, 116, 3603−3612, pag. 3609
B=0.0032849 # pm^(−1) (kg-solvent/mol-solute)^(1/2) -> From Table 2 of: J. Phys. Chem. B 2012, 116, 3603−3612, pag 3609
AdB=A/B # pm
a=348 # pm -> From Table 1 of: J. Phys. Chem. B 2012, 116, 3603−3612, pag. 3606
bp=120 # pm -> From Table 1 of: J. Phys. Chem. B 2012, 116, 3603−3612, pag. 3606
bm=362 # pm -> From Table 1 of: J. Phys. Chem. B 2012, 116, 3603−3612, pag. 3606 [CAPTION]
#kappa=B*np.sqrt(ionic_strength)

log_li=[]
log_cl=[]

for i in range (0, len(li)):
    kappa=B*li[i,0]
    f1=-AdB*(kappa/(1+kappa*a))
    f2= ( (2*np.exp(kappa*(a-bp))) - (kappa*(a-bp)) - 2) / (1+(kappa*bp))
    f3= ( (2*np.exp(kappa*(bm-a))) - (2*kappa*(bm-a)) - 2) / (1+(kappa*bm))
    log_li.append((f1 * ( 1 - f2 ) ))
    log_cl.append((f1 * ( 1 + f3 ) ))
l_li=np.array(log_li)
l_cl=np.array(log_cl)
```

```python
plt.figure(figsize=(6, 6), dpi= 80, facecolor='w', edgecolor='k')
plt.tick_params(axis='both', which='both', length=10.0, labelleft=True, labelbottom=True, labelsize=20.0)

#plt.plot(li[:,0], np.exp(dha[:]), linewidth=2.5, label="DH")
plt.plot(li[:,0], np.log10(li[:,1]), 'o', markerfacecolor='green', markersize=12, markeredgecolor='blue', 
        linewidth=2.5, label="Li+, Exp.")
plt.plot(li[:,0], l_li[:], linewidth=2.5, label="Li+, DH-SIS")
plt.plot(li[:,0], np.log10(cl[:,1]), 'o', markerfacecolor='blue', markersize=12, markeredgecolor='orange', 
        linewidth=2.5, label="Cl-, Exp.")
plt.plot(li[:,0], l_cl[:], linewidth=2.5, label="CL-, DH-SIS")
plt.plot((np.sqrt(ist[:])), (lg[:]), linewidth=2.5)
plt.plot((np.sqrt(ist[:])), (lg[:]), 'o', markerfacecolor='red', markersize=12, markeredgecolor='blue', 
         linewidth=2.5, label="DH")

plt.xlim(0,1.2)
plt.ylim(-0.6,0.2)

plt.xlabel('$\sqrt{m}$', fontsize=26)
plt.ylabel('$\log_{10}\gamma_{Li^{+},Cl^{-}}$', fontsize=26)
plt.legend()
plt.savefig('foo.png')
plt.show()
```
